```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import src.utils.metrics as metrics
import config.saveoursystem1 as save_our_system_config

CSV_PATHS = ["trials/saveoursystem1_trials_only_maintain.csv", "trials/saveoursystem1_trials_only_repair.csv", "trials/saveoursystem1_trials_only_replace.csv", "trials/saveoursystem1_trials_weighted_random.csv"]
```

```python
pd.read_csv(CSV_PATHS[0], index_col=False, sep=";")
```

```python
datas = []
for csv_path in CSV_PATHS:
    data = pd.read_csv(csv_path, index_col=False, sep=";")
    data["method"] = "_".join(csv_path.split("_")[2:4]).strip()
    data["method"] = data["method"].str.replace(".csv", "", regex=False)
    datas.append(data)
```

```python
merged_data = pd.concat(datas, ignore_index=True)
merged_data
```

```python
moves_cols = ["Move_1","Move_2","Move_3","Move_4","Move_5","Move_6","Move_7","Move_8","Move_9","Move_10"]
scores_cols = ["Score_1","Score_2","Score_3","Score_4","Score_5","Score_6","Score_7","Score_8","Score_9","Score_10"]
merged_data[scores_cols] = merged_data[scores_cols].astype(int)
max_score_index = merged_data[scores_cols].iloc[:, -1].idxmax()
```
